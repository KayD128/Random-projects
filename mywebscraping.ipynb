```python
products_title = []
pricing = []
desc = []
products_image = []
gend = []

for content in containers:
    for val in content.findAll("div", {"class": "masonry-item"}):
        prices = val.findAll("div", {"class": "qa-advert-price"})
        pricing.append(prices[0].text)
                
        name = val.findAll("div", {"class": "b-advert-title-inner qa-advert-title b-advert-title-inner--div"})
        products_title.append(name[0].text)
        
        description = val.findAll("div", {"class": "b-list-advert-base__description-text"})
        if len(description) == 0:
            desc.append(np.nan)
        else:
            desc.append(description[0].text)
        
        image = val.findAll("img")
        products_image.append(image[0]["src"])
        
        gender = val.findAll("div", {"class": "b-list-advert-base__item-attr"})
        if len(gender) == 0:
            gend.append(np.nan)
        else:
            gend.append(gender[0].text)
    
print(len(products_title))
print(len(products_image))
print(len(pricing))
print(len(desc))
print(len(gend))
```

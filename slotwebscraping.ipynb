```python
from selenium import webdriver  
from selenium.common.exceptions import NoSuchElementException  
from selenium.webdriver.common.keys import Keys  
from bs4 import BeautifulSoup as soup
# To run Chrome in head mode (this will open Chrome interface) 
# browser = webdriver.Chrome(executable_path=r"C:\Users\YEMI")  
browser = webdriver.Firefox()
browser.get('https://www.jumia.com.ng/catalog/?q=shoes&page=3#catalog-listing') 
# The driver.page_source will return the full page HTML code.
html_source = browser.page_source  
browser.quit()
```

```python
page_soup = soup(html_source,'html.parser')
container = page_soup.findAll('div', {'class':'-pvs col12'})  
# print(container)
print(len(container))
print(soup.prettify(container[0]))
```

```python
page=container[0].findAll("article", {"class": "prd _fb col c-prd"})
len(page)
```

```python
my_image=container[0].findAll('img', {"class":"img"})
my_image[0]["data-src"]
# print(soup.prettify(myPage[0]))
# print(soup.prettify(myPage[0]))
```

```python
my_name=container[0].findAll('h3', {"class":"name"})
my_name[0].text
```

```python
price_range=container[0].findAll('div', {"class":"prc"})
price_range[0].text
```

```python
image = []
name = []
price = []

for contents in page:
    my_image=contents.findAll('img', {"class":"img"})
    my_name=contents.findAll('h3', {"class":"name"})
    price_range=contents.findAll('div', {"class":"prc"})
    if len(my_image) != 0:
        images = my_image[0]["data-src"]
        image.append(images)
    if len(my_name) != 0:
        names = my_name[0].text
        name.append(names)    
    if len(price_range) != 0:
        prices = price_range[0].text
        price.append(prices)    

print(len(image))
print(len(name))
print(len(price))
```

```python
df = pd.DataFrame({"Product Name": name, "Product Image":image, "Price":price})
df.to_csv("jumiaproduct.csv")
jumia = pd.read_csv("jumiaproduct.csv")
jumia
```
